 ***importamos las bibliotecas necesarias***

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, norm
```

### Instructions

1. We will have another simple example on two sample t test (pooled- when the variances are equal). But this time this is a one sided t-test

In a packing plant, a machine packs cartons with jars. It is supposed that a new machine will pack faster on the average than the machine currently used. To test that hypothesis, the times it takes each machine to pack ten cartons are recorded. The results, in seconds, are shown in the tables in the file `files_for_lab/machine.txt`.
Assume that there is sufficient evidence to conduct the t test, does the data provide sufficient evidence to show if one machine is better than the other

# ***Paso 1: presentación de los datos***


```python
df= pd.DataFrame({'nueva':[42.1,41,41.3,41.8,42.4,42.8,43.2,42.3,41.8,42.7],'vieja':[42.7,43.6, 43.8,43.3,42.5, 43.5, 43.1, 41.7,44,44.1]})
mediavieja, medianueva=df['vieja'].mean(), df['nueva'].mean()
desvvieja, desvnueva=df['vieja'].std(), df['nueva'].std()
```


Como paso intermedio, para facilitar la visualización voy a graficar la distribución del Df 



```python
x=np.arange(40,44,0.1)

plt.figure(figsize=(6,4))
plt.plot(x,norm.pdf(x,mediavieja,desvvieja), label='vieja', color='k')
plt.plot(x,norm.pdf(x,medianueva,desvnueva), label='nueva', color='b')
plt.ylabel('producción')
plt.xlabel('tiempo de producción')
plt.title ('Distribución de la producción')

plt.vlines(mediavieja, ymin=0.0 , ymax=0.6 , color='k') #Aquí encontramos la media graficada
plt.vlines(medianueva, ymin=0.0 , ymax=0.6 , color='b') #Aquí encontramos la media graficada
plt.text(41.5, 0.6, '$\mu_{vieja}=$'f'${mediavieja.round(1)}$')
plt.text(42.5, 0.6, '$\mu_{nueva}=$'f'${medianueva.round(1)}$')
plt.legend()
```

# Paso 2: Two Sample T-test

 Con esta fórmula obtenemos el p_value a la hora de correrlo. si quisieramos el valor estadístico se pone 0 y si quisieramos ambos, no tendríamos que poner nada.

```python
p_value = ttest_ind(df['nueva'],df['vieja'])[1].round(5)
```

```python
p_value
```

# Conclusión:
Con los datos obtenido en el p_value (un resultado inferior a 0.05) nos hace pensar que si hay unadiferencia productiva a tener en cuenta entre las dos máquinas
